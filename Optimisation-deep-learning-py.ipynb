## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ["EIN", "NAME"])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ["T13", "T12", "T2", "T25", "T14", "T29", "T15", "T17"]

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df["CLASSIFICATION"].value_counts()[application_df["CLASSIFICATION"].value_counts() > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_stay = ["C1000", "C2000", "C1200", "C3000", "C2100", "C7000", "C1700", "C4000"]


# Replace in dataframe
for index, rows in application_df.iterrows():
  cls = rows["CLASSIFICATION"]
  if cls not in classifications_to_stay:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
# Creating all dummies for categorical data
application_dummies = pd.get_dummies(application_df[["APPLICATION_TYPE", "AFFILIATION", "CLASSIFICATION", "USE_CASE", 
                                         "ORGANIZATION", "INCOME_AMT", "SPECIAL_CONSIDERATIONS"]])
# merge into the main dataframe and drop the unwanted columns for fitting into model
application_dummies_df = pd.concat([application_df, application_dummies], axis = 1).drop(columns = ["APPLICATION_TYPE", "AFFILIATION", 
                                                                                                    "CLASSIFICATION", "USE_CASE", "ORGANIZATION", 
                                                                                                    "INCOME_AMT", "SPECIAL_CONSIDERATIONS"])
application_dummies_df.head()
```

```python
# Split our preprocessed data into our features (X) and target (y) arrays
# target (y) array = IS_SUCCESSFUL, others are features (X) arrays
X = application_dummies_df.drop(columns = ["IS_SUCCESSFUL"])
y = application_dummies_df["IS_SUCCESSFUL"]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state= 2)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units= 676, activation="relu", input_dim = 47))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units= 484, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units= 225, activation="relu"))

# Forth hidden layer
nn.add(tf.keras.layers.Dense(units= 64, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units= 1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = "binary_crossentropy", optimizer= "adam", metrics= ["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=250)
```

```python
 # Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
